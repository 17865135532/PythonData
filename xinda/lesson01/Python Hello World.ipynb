## 美国婴儿名字数据
* 数据来源： https://www.ssa.gov/oact/babynames/limits.html
* 包括1880至2016所有年份出生的婴儿名。
* 我们使用这个例子来展示python强大的数据总结和可视化功能，代码细节将在后续课程中讲解。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re as re
import scipy

%matplotlib inline
```

```python
# 读取数据

years = range(1880, 2017) # range函数：第三节课讲解
pieces = [] # 列表：第二节课讲解
columns = ['name', 'gender', 'frequency']

for year in years: # for循环：第二节课讲解
    path = './names/yob%d.txt' % year 
    frame = pd.read_csv(path, names=columns) # 数据读取： 第八节课讲解
    frame['year'] = year # 增加变量year：第八节课讲解
    pieces.append(frame)
```

```python
baby_names = pd.concat(pieces, ignore_index=True) # 转换为pd数据： 第五节课讲解
```

```python
baby_names
```

```python
# 总共189万行数据
# 查看前十行数据
# 注意：python默认的索引从0开始
# 第一行：Mary是个女性名字（gender为F），在1880年总共有7065个婴儿取名叫Mary
baby_names.head(10)
```

```python
baby_names.describe() # 使用describe方法查看数据基本信息
```

## 数据科学的一个特征是通过数据回答问题
### 哪些名字出现的频率最高？ 

```python
baby_names.groupby('name').agg({'frequency': sum})
```

```python
# James, John, Robert, Micheal, Mary...都是耳熟能详的名字
baby_names.groupby('name').agg({'frequency': sum}).sort_values(by=['frequency'], ascending=[0])
```

### 每年出生的男孩和女孩的个数分别是多少？

```python
# 使用pivot_table方法查看（第六节课讲解）
freq_by_gender_year = baby_names.pivot_table(index ='year', columns='gender',
                                             values = 'frequency', aggfunc=sum)
```

```python
# 使用tail方法查看最近几年出生人数
freq_by_gender_year.tail()
```

### 数据可视化往往是数据分析的第一步

```python
# 一行命令即可做出高质量图形
freq_by_gender_year.plot(title='Frequency by year and gender')
```

## 起名趋势分析

增加一个变量rank，这个是根据年份性别依据名字出现频率所产生的次序

```python
baby_names['ranked'] = baby_names.groupby(['year', 'gender'])['frequency'].rank(ascending=False)
```

```python
baby_names.head(10)
```

计算每个名每年按性别占总出生人数的百分比



```python
def add_pct(group):# 自定义函数，第三节课讲解
    group['pct'] = group.frequency / group.frequency.sum()*100
    return group

# groupby和apply函数在课程第二部分讲解
baby_names = baby_names.groupby(['year', 'gender']).apply(add_pct) 
```

```python
# 查看新加的百分比（pct）
baby_names.head()
```

### 首先查看每先查看每年最流行的名字所占百分比趋势

将数据分为男孩和女孩

```python
dff = baby_names[baby_names.gender == 'F'] #逻辑判断 ==， 第二节课讲解
dfm = baby_names[baby_names.gender == 'M']
```

获取每年排名第一的名字

```python
rank1m = dfm[dfm.ranked == 1]
rank1f = dff[dff.ranked == 1]
```
